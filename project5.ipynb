```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Read data from file
df = pd.read_csv('epa-sea-level.csv')

df.head()
```

```python
# Create scatter plot

fig,ax = plt.subplots(figsize=(12,7))

x = df['Year']
y = df['CSIRO Adjusted Sea Level']
plt.scatter(x,y)
```

```python
#Create first line of best fit
slope,intercept,r_value,p_value,std_err = linregress (df['Year'],df['CSIRO Adjusted Sea Level'])
s = slope
p = intercept

x_pred = pd.Series([i for i in range (1880,2050)])
y_pred = s*x_pred+p

ax = plt.plot(x_pred,y_pred)
```

```python
 # Create second line of best fit
best_df = df.loc[df['Year']>=2000]
slope1,intercept1,r_value1,p_value1,std_err1 = linregress(best_df['Year'],best_df['CSIRO Adjusted Sea Level'])
s1 = slope1
p1 = intercept1

x_pred1 = pd.Series([i for i in range (2000,2050)])
y_pred1 = s1*x_pred1+p1

ax = plt.plot(x_pred1 , y_pred1)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')


plt.savefig('sea-level-plot.png')
```
